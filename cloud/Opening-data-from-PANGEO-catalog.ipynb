### For the following to work, you need to be on a google cloud and be sure that intake, dask,  zarr, intake-xarray, gcsfs and aiohttp libraries are installed

```python
from intake import open_catalog
import zarr
```

#### Altimetry data : https://catalog.pangeo.io/browse/master/ocean/altimetry/

```python
cat = open_catalog("https://raw.githubusercontent.com/pangeo-data/pangeo-datastore/master/intake-catalogs/ocean/altimetry.yaml")
list(cat)
```

```python
dssat  = cat["al"].to_dask()
dssat
```
